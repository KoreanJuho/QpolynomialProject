```python
import numpy as np
```

```python
y = np.array([2.0, 1.3, 2.4, 1.5, 0.6, 2.0, 1.0, 2.0, 1.3, 0.9])
x1 = np.array([3.0, 1.1, 3.5, 2.5, 0.6, 2.8, 1.3, 3.3, 2.0, 1.0])
x2 = np.array([3.2, 3.0, 3.6, 2.6, 1.9, 3.5, 2.1, 3.4, 2.8, 2.3])
```

```python
A = np.array([
    [len(y), np.sum(x1), np.sum(x2)],
    [np.sum(x1), np.sum(x1**2), np.sum(x1*x2)],
    [np.sum(x2), np.sum(x1*x2), np.sum(x2**2)]
])

B = np.array([np.sum(y), np.sum(x1*y), np.sum(x2*y)])

Xc = np.linalg.solve(A, B)

print('Xc = ', Xc)
```
